```python
import h5py as h5
```

```python
f = h5.File("lines.h5", "r")
```

```python
nu = f["/nu"]
T = f["/T"]
```

```python
# Set burn in
i = arange(1e5)
burn_in = 50
i = i[burn_in:]
nu = nu[burn_in:]
T = T[burn_in:]
```

```python
dnu = loadtxt("data/spectral.dat", usecols=[0])
dT = loadtxt("data/spectral.dat", usecols=[1])
fi = exp(-(dnu - 37)**2 / (2*2**2))
sumdifi = (dT*fi).sum()
sumfi = (fi**2).sum()
```

```python
def t_post(T):
    return (1/T) * exp(T*sumdifi)*exp(-(T**2*sumfi)/2)
```

```python
t = arange(0.1, 3.5, 0.1)
plot(t, t_post(t))
```
